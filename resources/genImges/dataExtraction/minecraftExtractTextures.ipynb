```python
import os
import zipfile

def extract_zip_files(input_folder_path, output_folder_path):
    # Ensure the output folder exists, create if it doesn't
    if not os.path.exists(output_folder_path):
        os.makedirs(output_folder_path)

    # Iterate through all the files in the specified input folder
    for file_name in os.listdir(input_folder_path):
        # Check if the file is a zip file
        if file_name.endswith('.zip'):
            file_path = os.path.join(input_folder_path, file_name)
            # Create a separate output folder for each zip file
            extract_folder = os.path.join(output_folder_path, file_name[:-4])
            if not os.path.exists(extract_folder):
                os.makedirs(extract_folder)

            # Extract the zip file to its specific folder
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                print(f"Extracting {file_name} to {extract_folder}...")
                zip_ref.extractall(extract_folder)
                print(f"Done extracting {file_name}")

# Specify the input folder path and output folder path here
input_folder_path = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_MinecraftRaw"
output_folder_path = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_MinecraftRaw2"
extract_zip_files(input_folder_path, output_folder_path)
```

```python
import os
import shutil

def copy_files_from_subfolders(input_folder_path, output_root):
    # Funktion zum rekursiven Kopieren von Dateien
    def copy_files(src_path, dest_root):
        for subdir, dirs, files in os.walk(src_path):
            # Berechne den relativen Pfad ohne den "block" oder "blocks" Teil
            relative_path = os.path.relpath(subdir, src_path)
            # Erstelle den entsprechenden Zielordner
            dest_folder = os.path.join(dest_root, relative_path)
            os.makedirs(dest_folder, exist_ok=True)

            # Kopiere alle Dateien in den Zielordner
            for file in files:
                src_file = os.path.join(subdir, file)
                dest_file = os.path.join(dest_folder, file)
                shutil.copy(src_file, dest_file)

    # Gehe durch jeden Eintrag in der ersten Ebene des angegebenen Ordners
    for entry in os.listdir(input_folder_path):
        full_entry_path = os.path.join(input_folder_path, entry)

        # Prüfe, ob der Eintrag ein Ordner ist
        if os.path.isdir(full_entry_path):
            # Bestimme die Pfade für die gesuchten Unterordner
            blocks_path = os.path.join(full_entry_path, "assets\\minecraft\\textures\\blocks")
            block_path = os.path.join(full_entry_path, "assets\\minecraft\\textures\\block")

            # Wähle den existierenden Pfad aus und kopiere die Dateien
            if os.path.exists(blocks_path):
                dest_folder = os.path.join(output_root, entry)
                copy_files(blocks_path, dest_folder)
            if os.path.exists(block_path):
                dest_folder = os.path.join(output_root, entry)
                copy_files(block_path, dest_folder)

# Pfad zum übergeordneten Verzeichnis
input_folder_path = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_MinecraftRaw2"
output_root = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_MinecraftRaw3"
copy_files_from_subfolders(input_folder_path, output_root)
```

```python
import os
import shutil

# Pfad zur Textdatei, die die Dateinamen enthält
file_list_path = "minecraftBlockDateiliste.txt"

# Wurzelverzeichnis, in dem die Dateien gesucht werden
source_root = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_MinecraftRaw3"

# Zielverzeichnis, in das die Dateien kopiert werden sollen
destination_root = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_Minecraft"

# Liste der Dateinamen aus der Textdatei einlesen
with open(file_list_path, 'r') as file:
    file_names = [line.strip() for line in file]

# Durchsucht alle Unterordner und kopiert die gefundene Datei
for subdir, dirs, files in os.walk(source_root):
    for file in files:
        if file in file_names:
            # Ermittelt den relativen Pfad im Quellverzeichnis
            relative_path = os.path.relpath(subdir, source_root)

            # Erstellt den Zielordnerpfad
            destination_path = os.path.join(destination_root, relative_path)

            # Erstellt den Zielordner, falls er nicht existiert
            os.makedirs(destination_path, exist_ok=True)

            # Kopiert die Datei
            shutil.copy2(os.path.join(subdir, file), destination_path)

print("Kopieren abgeschlossen.")
```

```python
import os
from PIL import Image

# Pfad zum Wurzelverzeichnis, in dem die Bilder gesucht werden
root_directory = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_Minecraft"

# Mindestgröße, unterhalb derer die Bilder gelöscht werden sollen
min_size = (128, 128)

def search_and_delete(directory):
    files_to_delete = []

    for entry in os.scandir(directory):
        if entry.is_file() and entry.name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            try:
                with Image.open(entry.path) as img:
                    if img.size < min_size:
                        files_to_delete.append(entry.path)

            except IOError:
                print(f"Fehler beim Öffnen der Datei {entry.name}")
        
        elif entry.is_dir():
            # Rekursiver Aufruf für Unterordner
            search_and_delete(entry.path)

    # Löschen der Dateien nach dem Durchlaufen des Verzeichnisses
    for file in files_to_delete:
        os.remove(file)
        print(f"Gelöscht: {file}")

# Start der Suche und Löschaktion im Wurzelverzeichnis
search_and_delete(root_directory)

print("Löschvorgang abgeschlossen.")
```

```python
import os
from PIL import Image

# Pfad zum Wurzelverzeichnis, in dem die Bilder gesucht werden
root_directory = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_Minecraft"

# Funktion, um zu überprüfen, ob eine Zahl eine Potenz von 2 ist
def is_power_of_two(n):
    return (n != 0) and (n & (n - 1) == 0)

# Funktion zur Durchsuchung eines Verzeichnisses und seiner Unterordner
def search_and_delete(directory):
    files_to_delete = []

    for entry in os.scandir(directory):
        if entry.is_file() and entry.name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            try:
                with Image.open(entry.path) as img:
                    if img.width != img.height or not ((is_power_of_two(img.width) and is_power_of_two(img.height))):
                        files_to_delete.append(entry.path)

            except IOError:
                print(f"Fehler beim Öffnen der Datei {entry.name}")
        
        elif entry.is_dir():
            # Rekursiver Aufruf für Unterordner
            search_and_delete(entry.path)

    # Löschen der Dateien nach dem Durchlaufen des Verzeichnisses
    for file in files_to_delete:
        os.remove(file)
        print(f"Gelöscht: {file}")

# Start der Suche und Löschaktion im Wurzelverzeichnis
search_and_delete(root_directory)

print("Löschvorgang abgeschlossen.")
```

```python
import os
from PIL import Image
from collections import defaultdict

# Pfad zum Wurzelverzeichnis, in dem die Bilder gesucht werden
root_directory = "C:/Users/Dennis/Desktop/Pro/AITest/imgGen/Data_Minecraft"

# Ein defaultdict zum Speichern der Zählung der Bildgrößen
size_count = defaultdict(int)

# Funktion zur Durchsuchung eines Verzeichnisses und seiner Unterordner
def count_image_sizes(directory):
    for entry in os.scandir(directory):
        if entry.is_file() and entry.name.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
            try:
                with Image.open(entry.path) as img:
                    size_count[img.size] += 1

            except IOError:
                print(f"Fehler beim Öffnen der Datei {entry.name}")
        
        elif entry.is_dir():
            # Rekursiver Aufruf für Unterordner
            count_image_sizes(entry.path)

# Start der Suche im Wurzelverzeichnis
count_image_sizes(root_directory)

# Ausgabe der Ergebnisse
for size, count in size_count.items():
    print(f"Größe {size}: {count} Bild(er)")

print("Auswertung abgeschlossen.")
```
