```python
import os

def remove_specific_files(directory):
    for filename in os.listdir(directory):
        if "unity.zip" in filename.lower() or "-ue.zip" in filename.lower():
            file_path = os.path.join(directory, filename)
            os.remove(file_path)
            print(f"Removed: {filename}")

# Ersetzen Sie 'path_to_directory' mit dem Pfad des Ordners, aus dem Sie Dateien entfernen möchten
directory_path = 'freepbr'
remove_specific_files(directory_path)
```

```python
import os
import zipfile

def extract_all_zip_files(directory):
    for filename in os.listdir(directory):
        if filename.endswith(".zip"):
            file_path = os.path.join(directory, filename)
            with zipfile.ZipFile(file_path, 'r') as zip_ref:
                # Erstellen eines Unterverzeichnisses für jede ZIP-Datei
                extract_path = os.path.join(directory, os.path.splitext(filename)[0])
                os.makedirs(extract_path, exist_ok=True)
                zip_ref.extractall(extract_path)
                print(f"Extracted: {filename} to {extract_path}")

# Ersetzen Sie 'path_to_directory' mit dem Pfad des Ordners, dessen ZIP-Dateien Sie entpacken möchten
directory_path = 'freepbr'
extract_all_zip_files(directory_path)
```

```python
import os
import shutil

def flatten_and_copy_files(src_directory, dest_directory):
    for root, _, files in os.walk(src_directory):
        for file in files:
            src_path = os.path.join(root, file)
            dest_path = os.path.join(dest_directory, file)
            # Überprüfung auf Dateinamenskonflikte
            if os.path.exists(dest_path):
                print(f"Konflikt: {dest_path} existiert bereits. Datei nicht kopiert: {src_path}")
                continue
            shutil.copy2(src_path, dest_path)
            print(f"Copied: {src_path} to {dest_path}")

# Ersetzen Sie diese Pfade mit dem Quell- und Zielverzeichnis
src_directory_path = 'freepbr'
dest_directory_path = 'freepbrEXTRACTED'
os.makedirs(dest_directory_path, exist_ok=True)
flatten_and_copy_files(src_directory_path, dest_directory_path)
```

```python
import os

def delete_files_with_keywords(directory, keywords):
    for filename in os.listdir(directory):
        if any(keyword.lower() in filename.lower() for keyword in keywords):
            file_path = os.path.join(directory, filename)
            os.remove(file_path)
            print(f"Deleted: {filename}")

# Ersetzen Sie 'path_to_directory' mit dem Pfad des Ordners, aus dem Sie Dateien löschen möchten
directory_path = 'freepbrEXTRACTED'
# Liste der Keywords, die zur Identifikation der zu löschenden Dateien verwendet werden sollen
keywords = ["preview", ".obj", "-ogl.png", "ao.png", "height.png", "roughness.png", "normal", "metallic"]
delete_files_with_keywords(directory_path, keywords)
```

```python
import os
from PIL import Image
from collections import Counter

def count_image_size_overlaps(directory):
    image_files = [f for f in os.listdir(directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]
    image_dimensions = []

    for file in image_files:
        path = os.path.join(directory, file)
        with Image.open(path) as img:
            width, height = img.size
            image_dimensions.append((width, height))

    dimension_counts = Counter(image_dimensions)
    return dimension_counts

# Setze hier den Pfad deines Ordners
directory = 'Data_freepbr'
dimension_overlaps = count_image_size_overlaps(directory)

# Zum Ausgeben der Ergebnisse
for dimensions, count in dimension_overlaps.items():
    if count > 1:
        print(f"Größe {dimensions} kommt {count} Mal vor.")

```

```python
import os
from PIL import Image

def resize_images_to_1024x1024(directory, directory_save):
    image_files = [f for f in os.listdir(directory) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp'))]

    for file in image_files:
        path = os.path.join(directory, file)

        #create Folder if not exists
        if not os.path.exists(directory_save):
            os.makedirs(directory_save)

        with Image.open(path) as img:
            img_resized = img.resize((1024, 1024))
            img_resized.save(directory_save + file)

# Setze hier den Pfad deines Ordners
directory = 'Data_freepbr'
directory_save = 'Data_freepbr_resized/'
resize_images_to_1024x1024(directory, directory_save)
```
