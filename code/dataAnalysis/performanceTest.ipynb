```python
import matplotlib.pyplot as plt

batch_sizes = [1, 2, 4, 8, 16, 24, 32, 40, 48, 56, 64]
avg_times_per_token = [0.01886, 0.00950, 0.00519, 0.00353, 0.00282, 0.00264, 0.00257, 0.00251, 0.0042, 0.00590, 0.00719]

plt.figure(figsize=(10, 5))
plt.plot(batch_sizes, avg_times_per_token, marker='o', linestyle='-')
plt.title('Average Time Per Token vs. Batch Size for CIT Model [4] 512 Token')
plt.xlabel('Batch Size')
plt.ylabel('Average Time Per Token (sec)')
plt.grid(True)

# Define x-ticks normally
plt.xticks(batch_sizes)

# Modify y-ticks to exclude specific labels
ytick_values = sorted(set(avg_times_per_token))
ytick_labels = [f"{value:.5f}" if value not in [0.00282, 0.00264, 0.00257] else "" for value in ytick_values]
plt.yticks(ytick_values, ytick_labels)

plt.savefig('../../latex/imgs/performanceTest.png')
plt.show()


```
